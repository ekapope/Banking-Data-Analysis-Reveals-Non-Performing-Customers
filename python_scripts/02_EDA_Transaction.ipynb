```python
# Importing the libraries

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt 
import matplotlib 
import seaborn as sns
from collections import Counter
from matplotlib import pyplot
```

```python
#Importing the basetable
basetable = pd.read_csv('C:/Users/sakhtar/Downloads/Financial Programming/Group/basetable.csv', sep=',')
```

```python
# Plot - Unknown Transaction Amount Frequency

f = plt.figure(1)
y = basetable['mean_trans_amount_of_Unknown_Transaction_Type']
x = basetable['nbr_trans_Unknown_Transaction_Type']
#y_pos = np.arange(len(x))
plt.xlim(1, 54)
plt.bar(x, y, color=(0.2, 0.4, 0.6, 0.8))
# Custom Axis title
plt.xlabel('Number of Unknown transactions', fontweight='bold', color = 'green', fontsize='10', horizontalalignment='center')
plt.ylabel('Average Transaction Amount', fontweight='bold', color = 'red', fontsize='10', horizontalalignment='center')
plt.title('Unknown Transaction Frequency')
plt.savefig('fig_'+str('Unknown Transaction Frequency')+'.png', dpi=200,bbox_inches="tight")
plt.show()
```

```python
#drop na values in column '1yr_mean_trans_amount_by_account_id'
basetable = basetable.dropna(subset=['1yr_mean_trans_amount_by_account_id'])
```

```python
# Plot - Average Transaction Amount in the last 1 year

sns.distplot(basetable['1yr_mean_trans_amount_by_account_id'])
#sns.plt.show()
plt.xlabel('Mean Transaction amount in last 1 year', fontweight='bold', color = 'green', fontsize='10', horizontalalignment='center')
plt.title('Average Transaction Amount in the last 1 year')
plt.savefig('fig_'+str('Average Transaction Amount in the last 1 year')+'.png', dpi=200,bbox_inches="tight")
#plt.show()
```

```python
#drop na values in column '3yr_mean_trans_amount_by_account_id'
basetable = basetable.dropna(subset=['3yr_mean_trans_amount_by_account_id'])
```

```python
# Plot - Average Transaction Amount in the last 3 years
sns.distplot(basetable['3yr_mean_trans_amount_by_account_id'])
#sns.plt.show()
plt.xlabel('Mean Transaction amount in last 3 years', fontweight='bold', color = 'green', fontsize='10', horizontalalignment='center')
plt.title('Average Transaction Amount in the last 3 years')
plt.savefig('fig_'+str('Average Transaction Amount in the last 3 years')+'.png', dpi=200,bbox_inches="tight")
#plt.show()
```

```python
# Plot - Client Age and Total Amount transacted

plt.scatter(basetable['client_age'], basetable['alltime_sum_trans_amount_by_account_id'], color='y',alpha=0.5)
plt.xlabel('Client Age')
plt.ylabel('Transaction Amount per account id')
plt.title('Client Age and Total Amount transacted')
#plt.savefig('fig_'+str('Client Age and Total Amount transacted')+'.png', dpi=200,bbox_inches="tight")
plt.show()
```

```python
# Plot - Client Age and Account type used

f = plt.figure(1)
y = basetable['client_age']
x = basetable['type']
plt.xlabel('Account Type used')
plt.ylabel('Client Age')
#y_pos = np.arange(len(x))
#plt.xlim(1, 155)
plt.bar(x, y,color='m',alpha=0.3)
plt.title('Client Age and Account Type used')
plt.savefig('fig_'+str('Client Age and Account type used')+'.png', dpi=200,bbox_inches="tight")
plt.show()
```
